### Multicollinearity

   in a model that has been built using several independent variables, some of the features might be interrelated.As a result the presence of that variable in the model is redundant. We need to remove this columns in order to remove the Multicollinearity. However note that though it is a bit silly to have both the variables in the model when the are highly correlated but that doesn’t have adverse effects on prediction.

  Handling multicollinearity is not mandatory depending on the purpose and the model. For example, tree based models such as decision tree, bagging, random forest, and boosting are robust to multicollinearity issues in their aspect of prediction accuracy
    
    -- There are two ways it can be handled.
     --- Correlation Matrix
     --- Variance Inflation Factor (VIF)
     
   This technique holds good when there is less number of features are present in the dataset. However, when we have very high number of features, then we can use Ridge and Lasso regression to penalize thoise features by lambda value

### Rescaling the features
---   Standardization vs Normalization

Rescalling of the numeric features are extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients.This might become very annoying at the time of model evaluation. 

One more reason of Rescalling is, bringing all the variables in 0 to 1 range improves the optimization of the model. In other words, the backend gradient descent algorithm that tries to reduce the cost function, become faster

There are two ways Rescalling can be done. They are:

     --Standardization (mean - 0; sigma -1)
     --Normalization (Min-Max Scalling)

#### Normalization:
  - If x is a feature then normalization is (x-xmin)/(xmax-xmin).
  - It converts the data between 0 to 1
#### Standardization: 
  - If x is a feature then normalization is (x-mu)/(sigma). (x−μ)/σ

The advantage of Standardisation over the other is that it doesn't compress the data between a particular range as in Min-Max scaling. This is useful, especially if there is are extreme data point (outlier)
